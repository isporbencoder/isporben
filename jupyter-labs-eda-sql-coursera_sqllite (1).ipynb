## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
conn = sqlite3.connect(':memory:')  # in memory database
df.to_sql(name="spacexdata", con=conn, if_exists="replace")

q = pd.read_sql('select * from spacexdata', conn)
q
```

```python
q = pd.read_sql('select distinct Launch_Site from spacexdata', conn)
q
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
q = pd.read_sql("select * from spacexdata where Launch_Site like 'CCA%' limit 20", conn)
q
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
q = pd.read_sql("select sum(PAYLOAD_MASS__KG_) from spacexdata where Customer='NASA (CRS)'", conn)
q
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
q = pd.read_sql("select avg(PAYLOAD_MASS__KG_) FROM spacexdata where Booster_Version='F9 v1.1'", conn)
q
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
q = pd.read_sql("select min(Date)from spacexdata where Landing_Outcome='Success (ground pad)'", conn)
q
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
q = pd.read_sql("select DISTINCT Booster_Version FROM spacexdata where Landing_Outcome ='Success (drone ship)' and PAYLOAD_MASS__KG_ between 4000 and 6000", conn)
q
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
q = pd.read_sql("select count(Mission_Outcome)from spacexdata where Mission_Outcome = 'Success' or Mission_Outcome = 'failure'", conn)
q
#%sql select count(MISSION_OUTCOME) from SPACEXTBL where 
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
q = pd.read_sql('SELECT Booster_Version from spacexdata where PAYLOAD_MASS__KG_ = (select max (PAYLOAD_MASS__KG_) from spacexdata)', conn)
q

```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
q = pd.read_sql("SELECT substr(Date, 6, 2) as month, DATE, BOOSTER_VERSION, LAUNCH_SITE, [Landing_Outcome] FROM spacexdata WHERE [Landing_Outcome] = 'Failure (drone ship)' AND substr(Date, 0, 5) = '2015'", conn)
q
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
q = pd.read_sql("""
    SELECT Landing_Outcome, COUNT(*) as total_count
    FROM spacexdata
    WHERE Date BETWEEN '2010-06-04' AND '2017-03-20'
    GROUP BY Landing_Outcome
    ORDER BY total_count DESC
""", conn)
q
```
